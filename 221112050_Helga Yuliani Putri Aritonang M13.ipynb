### Tutorial-02: Permen

Terdapat sekantong permen dengan lima jenis permen, yaitu 'Kit Kat', 'Snickers', 'Milky Way', 'Toblerone', dan 'Twix'. Gunakan jenis visualisasi lingkaran untuk menunjukkan persentase peluang bahwa kita akan mengeluarkan permen Snickers dari kantong jika dilakukan pengambilan secara acak.


```python
import matplotlib.pyplot as plt

nama = ['Kit Kat', 'Snickers', 'Milky Way', 'Toblerone', 'Twix']
jumlah = [52, 39, 78, 13, 78]

indeks_snickers = nama.index('Snickers')
peluang_snickers = jumlah[indeks_snickers] / sum(jumlah) * 100

warna = ['#8B4513', '#FFE4B5', '#87AE73', 'orange', '#C41E3A']

explode = [0 if i != indeks_snickers else 0.08 for i in range(len(nama))]

plt.figure(figsize=(5, 5))
plt.pie(jumlah, labels=nama, autopct='%1.1f%%', startangle=0, colors=warna, explode=explode)
plt.title(f'Peluang Mengeluarkan Snickers: {peluang_snickers:.1f}%')
plt.show()
```

### Tutorial-03: Makanan

Restoran memiliki sisa menu makanan penutup yang terlalu banyak. Mereka ingin menghapus beberapa item dari menu. Agar sebagian besar pelanggan senang, maka restoran ingin menghapus hanya tiga makanan penutup yang paling tidak populer dari menu. Restoran memiliki data berapa kali makanan penutup tersebut dijual dalam seminggu terakhir. Gunakan jenis visualisasi batang untuk menunjukkan popularitas relatif dari makanan penutup.


```python
import matplotlib.pyplot as plt

penjualan_dessert = {
    'Lava Cake': 14,
    'Mousse': 5,
    'Chocolate Cake': 12,
    'Ice Cream': 19,
    'Truffles': 6,
    'Brownie': 8,
    'Chocolate Chip Cookie': 12,
    'Chocolate Pudding': 9,
    'Souffle': 10,
    'Chocolate Cheesecake': 17,
    'Chocolate Chips': 2,
    'Fudge': 9,
    'Mochi': 13
}

sorted_penjualan = dict(sorted(penjualan_dessert.items(), key=lambda item: item[1], reverse=True))
makanan_yang_akan_dihapus = list(sorted_penjualan.keys())[-3:]

plt.figure(figsize=(6, 4))
bars = plt.bar(sorted_penjualan.keys(), sorted_penjualan.values(),width=0.5, color=['blue' if x not in makanan_yang_akan_dihapus else 'red' for x in sorted_penjualan.keys()], label='Jumlah')

plt.legend()
plt.title('Popularitas Makanan Penutup')
plt.xlabel('Dessert')
plt.ylabel('Kuantiti Terjual')
plt.xticks(rotation='vertical')
plt.show()
```
