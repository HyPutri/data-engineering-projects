# Data Exploring and Analysis

### Tutorial-01:

- Langkah 1: Buat data frame bernama df yang berisi dictionary serta memiliki label sebagai berikut.

- Langkah 2: Tampilkan ringkasan informasi dasar dari data frame tersebut.
- Langkah 3: Tampilkan tiga baris pertama dari data frame tersebut.
- Langkah 4: Tampilkan hanya kolom Animal dan Age dari data frame tersebut.

```python
import pandas as pd

data = {
    'Animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'Age': [2.5, 3.0, 0.5, None, 5.0, 2.0, 4.5, None, 7.0, 3.0],
    'Priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no'],
    'Visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1]
}

df = pd.DataFrame(data)

#Tampilkan ringkasan informasi dasar
print("Langkah 2: Ringkasan Informasi Dasar")
print(df.info())

#Tampilkan tiga baris pertama
print("\nLangkah 3: Tiga Baris Pertama")
print(df.head(3))

#Tampilkan hanya kolom Animal dan Age
print("\nLangkah 4: Kolom Animal dan Age")
print(df[['Animal', 'Age']])
```

### Tutorial-02:

- Langkah 1: Buat data frame untuk menyimpan data yang dibaca dari file shopping_data.csv.
- Langkah 2: Untuk memastikan data terbaca dengan baik, tampilkan sebagian data (sebanyak 20 data) dari data frame tersebut.
- Langkah 3: Tampilkan nama-nama kolom yang tersedia dari data frame tersebut.
- Langkah 4: Akses data dari kolom umur pada data frame tersebut.
- Langkah 5: Tampilkan nilai umur dari data kelima yang ada di data frame.
- Langkah 6: Tampilkan data dari baris ke-5 sampai 20 dari kolom umur.
- Langkah 7: Tampilkan informasi statistik dari data frame tersebut dengan mengabaikan data nonnumerik.
- Langkah 8: Lakukan pengecekan apakah terdapat nilai Null pada data frame tersebut.
- Langkah 9: Tampilkan nilai mean dari masing-masing kolom pada data frame tersebut.
- Langkah 10: Isi nilai yang kosong pada data frame tersebut dengan menggunakan nilai rata-rata dan tampilkan hasilnya. 

```python
# Langkah 1
import pandas as pd

df = pd.read_csv('shopping_data.csv')

# Langkah 2
print("\nLangkah 2: Sebagian data (20 data)")
print(df.head(20))

# Langkah 3
print("\nLangkah 3: Nama-nama kolom")
print(df.columns)

# Langkah 4
print("\nLangkah 4: Data dari kolom umur")
print(df['Age'])

# Langkah 5
print("\nLangkah 5: Nilai umur dari data kelima")
print(df['Age'].iloc[4])

# Langkah 6
print("\nLangkah 6: Data dari baris ke-5 sampai 20 dari kolom umur")
print(df['Age'].iloc[4:20])

# Langkah 7
print("\nLangkah 7: Informasi statistik (numerik)")
print(df.describe())

# Langkah 8
print("\nLangkah 8: Pengecekan nilai Null")
print(df.isnull().any())

# Langkah 9
print("\nLangkah 9: Nilai mean dari masing-masing kolom")
print(df.select_dtypes(include='number').mean())

# Langkah 10
df_filled = df.fillna(df.select_dtypes(include='number').mean())
print("\nLangkah 10: Data frame setelah mengisi nilai kosong")
print(df_filled)
```
