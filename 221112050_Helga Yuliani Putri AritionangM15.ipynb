Dalam melakukan analisis data gaji karyawan pada perusahaan ABC berdasarkan dataset Salaries.csv, kita dapat memanfaatkan berbagai teknik visualisasi data untuk mendapatkan wawasan yang mendalam. Melalui serangkaian pertanyaan dan tugas, kita dapat mengeksplorasi distribusi gaji, membandingkan perbedaan gaji antara pria dan wanita, menganalisis hubungan antara pengalaman dan gaji, serta mengidentifikasi tren dan pola menarik dalam dataset tersebut.

Langkah-langkah yang perlu dilakukan adalah sebagai berikut:

1. Eksplorasi Umum:
   - Visualisasikan distribusi gaji menggunakan histogram.
   - Bandingkan jumlah pria dan wanita dalam dataset menggunakan bar plot.
    
2. Analisis Gaji Berdasarkan Jenis Kelamin:
   - Buat boxplot untuk melihat distribusi gaji berdasarkan jenis kelamin.
   - Hitung rata-rata gaji untuk pria dan wanita, dan tampilkan dalam bar plot.

3. Pendidikan dan Gaji:
   - Hitung rata-rata gaji untuk setiap tingkat pendidikan ("rank") dan visualisasikan hasilnya.
   - Bandingkan distribusi gaji antara dosen dan profesor menggunakan box plot.

4. Hubungan Antara Pengalaman dan Gaji:
   - Buat scatter plot antara jumlah tahun pelayanan ("service") dengan gaji.
   - Tambahkan garis regresi pada scatter plot untuk melihat tren.

5. Eksplorasi Tren:
   - Hitung jumlah dosen dan profesor untuk setiap tahun layanan dan tampilkan dalam line plot.
   - Visualisasikan tren kenaikan atau penurunan gaji dari tahun ke tahun menggunakan line plot.

6. Distribusi Gaji Berdasarkan Disiplin:
   - Bandingkan distribusi gaji antara disiplin ilmu ("discipline") dengan menggunakan boxplot.
   - Hitung rata-rata gaji untuk masing-masing disiplin dan tampilkan dalam bar plot.

7. Eksplorasi Outlier:
   - Identifikasi dan visualisasikan outlier dalam dataset Salaries menggunakan box plot atau scatter plot.

8. Pengelompokan Gaji Berdasarkan Pendidikan dan Jenis Kelamin:
   - Gunakan seaborn untuk membuat heatmap yang menunjukkan rata-rata gaji berdasarkan tingkat pendidikan dan jenis kelamin.

9. Pengelompokan Gaji Berdasarkan Kategori Umur:
   - Buat histogram untuk menunjukkan distribusi umur karyawan.
   - Gunakan seaborn untuk membuat bar plot yang mengelompokkan gaji berdasarkan rentang umur.

10. Eksplorasi Korelasi:
    - Hitung korelasi antara variabel-variabel seperti "phd", "service", dan "salary".
    - Visualisasikan korelasi menggunakan heatmap.

```python
# Import library yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Baca dataset "Salaries.csv"
salaries_data = pd.read_excel("Salaries.xlsx")

# 1. Eksplorasi Umum
plt.hist(salaries_data['salary'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribusi Gaji')
plt.xlabel('Gaji')
plt.ylabel('Frekuensi')
plt.show()

sns.countplot(x='sex', data=salaries_data, palette='pastel')
plt.title('Jumlah Pria dan Wanita dalam Dataset')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Jumlah')
plt.show()
```

```python
sns.boxplot(x='sex', y='salary', data=salaries_data, palette='pastel')
plt.title('Distribusi Gaji Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Gaji')
plt.show()

average_salary_by_gender = salaries_data.groupby('sex')['salary'].mean().reset_index()
sns.barplot(x='sex', y='salary', data=average_salary_by_gender, palette='viridis')
plt.title('Rata-rata Gaji Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Rata-rata Gaji')
plt.show()
```

```python
# 3. Pendidikan dan Gaji
average_salary_by_rank = salaries_data.groupby('rank')['salary'].mean().reset_index()
sns.barplot(x='rank', y='salary', data=average_salary_by_rank, palette='viridis')
plt.title('Rata-rata Gaji Berdasarkan Tingkat Pendidikan')
plt.xlabel('Tingkat Pendidikan')
plt.ylabel('Rata-rata Gaji')
plt.show()

sns.boxplot(x='rank', y='salary', data=salaries_data, palette='viridis')
plt.title('Distribusi Gaji Antara Dosen dan Profesor')
plt.xlabel('Tingkat Pendidikan')
plt.ylabel('Gaji')
plt.show()
```

```python
# 4. Hubungan Antara Pengalaman dan Gaji

sns.scatterplot(x='service', y='salary', data=salaries_data, color='green')
plt.title('Hubungan Antara Jumlah Tahun Pelayanan dan Gaji')
plt.xlabel('Jumlah Tahun Pelayanan')
plt.ylabel('Gaji')
plt.show

sns.regplot(x='service', y='salary', data=salaries_data, scatter_kws={'s':10}, line_kws={'color':'blue'})
plt.title('Hubungan Antara Jumlah Tahun Pelayanan dengan Garis Regresi')
plt.xlabel('Jumlah Tahun Pelayanan')
plt.ylabel('Gaji')
plt.show
```

```python
# 5. Eksplorasi Tren
salaries_data['year'] = pd.DatetimeIndex(salaries_data['service']).year

sns.countplot(x='year', hue='rank', data=salaries_data, palette='viridis')
plt.title('Jumlah Dosen dan Profesor Tiap Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.show()

sns.lineplot(x='year', y='salary', data=salaries_data, hue='rank', palette='pastel')
plt.title('Tren Kenaikan atau Penurunan Gaji Tiap Tahun')
plt.xlabel('Tahun')
plt.ylabel('Gaji')
plt.show()
```

```python
# 6. Distribusi Gaji Berdasarkan Disiplin
sns.boxplot(x='discipline', y='salary', data=salaries_data, palette='viridis')
plt.title('Distribusi Gaji Berdasarkan Disiplin')
plt.xlabel('Disiplin')
plt.ylabel('Gaji')
plt.show()

average_salary_by_discipline = salaries_data.groupby('discipline')['salary'].mean().reset_index()
sns.barplot(x='discipline', y='salary', data=average_salary_by_discipline, palette='viridis')
plt.title('Rata-rata Gaji Berdasarkan Disiplin')
plt.xlabel('Disiplin')
plt.ylabel('Rata-rata Gaji')
plt.show()
```

```python
# 7. Eksplorasi Outlier
sns.boxplot(x='rank', y='salary', data=salaries_data, palette='viridis')
plt.title('Distribusi Gaji Antara Dosen dan Profesor (termasuk Outlier)')
plt.xlabel('Tingkat Pendidikan')
plt.ylabel('Gaji')
plt.show()
```

```python
# 8. Pengelompokan Gaji Berdasarkan Pendidikan dan Jenis Kelamin
heatmap_data = salaries_data.pivot_table(values='salary', index='sex', columns='rank', aggfunc='mean')
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".0f", cbar_kws={'label': 'Rata-rata Gaji'})
plt.title('Pengelompokan Gaji Berdasarkan Pendidikan dan Jenis Kelamin')
plt.xlabel('Tingkat Pendidikan')
plt.ylabel('Jenis Kelamin')
plt.show()
```

```python
# 9. Pengelompokan Gaji Berdasarkan Kategori Service
sns.histplot(data=salaries_data, x='service', hue='rank', multiple='stack', bins=20, palette='viridis')
plt.title('Distribusi Gaji Berdasarkan Rentang Umur (Berdasarkan Jumlah Tahun Pelayanan)')
plt.xlabel('Jumlah Tahun Pelayanan')
plt.ylabel('Jumlah')
plt.show()

plt.figure(figsize=(8, 4))
sns.barplot(data=salaries_data, x='service', y='salary', hue='rank', palette='viridis')
plt.title('Gaji Berdasarkan Rentang Umur')
plt.xlabel('Jumlah Tahun Pelayanan')
plt.ylabel('Gaji')
plt.show()
```

```python
# 10. Eksplorasi Korelasi
correlation_data = salaries_data[['phd', 'service', 'salary']].corr()

sns.heatmap(correlation_data, annot=True, cmap='YlGnBu', fmt='.2f', linewidths=.5)
plt.title('Heatmap Korelasi antara phd, service, dan salary')
plt.show()
```
